```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

netflix_data = pd.read_csv("netflix_titles.csv")
netflix_data.head(10)
```

```python
netflix_data.describe(include = 'all')
```

```python
netflix_data['type'].value_counts()
```

```python
ax = sns.countplot(x="type", data=netflix_data, palette='Set1')
```

It is evident that there are more movies on Netflix, than TV shows.

```python
print("Mean release year: ", round(netflix_data['release_year'].mean()))
print("Median release year: ", round(netflix_data['release_year'].median()))
print("Release year std: ", round(netflix_data['release_year'].std()))
```

```python
pd.crosstab(netflix_data['release_year'],netflix_data['type'], margins=True)
```

Here you can see how much content was published in a particular year

```python
plt.subplot(2,1,1)
ano = netflix_data['release_year'].value_counts()[:20].plot.bar(figsize=(18,10))
plt.xticks(rotation = 0)
plt.title('Most popular age of movies and TV shows on Netflix', pad = 15, fontsize = 18)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Amount',fontsize = 16)
plt.show()
```

This bar chart shows that the most material was published in 2018. Unexpectedly, 2021 ranks 6th in the top.

```python
top_countries = netflix_data['country'].value_counts().head(10)
fig = px.pie(names=top_countries.index, values=top_countries.values, labels={'index': 'Country'}, title='Countries that released the most movies and TV shows.')
fig.update_layout(title_x=0.45)
fig.update_traces(textposition='inside', textinfo='label+percent')
fig.show()
```

On this diagram we can see that the USA released the most content on Netflix as was expected. However, India takes second place that is quite surprisingly.

```python
listed_in = []
for i in range(len(netflix_data)):
    listed_in.extend(netflix_data.listed_in.iloc[i].split(','))

listed_dic = {}
for i in listed_in:
    listed_dic[i] = listed_in.count(i)

listed_dic = sorted(listed_dic.items(), key=lambda item: item[1], reverse=True)
listed_dic = dict(listed_dic)

height = list(listed_dic.values())[:15]
bars = list(listed_dic.keys())[:15]
y_pos = np.arange(len(bars))
plt.bar(y_pos, height)
plt.xticks(y_pos, bars)
plt.xlabel('Categories')
plt.xticks(rotation=90)
plt.ylabel('Number of content')
plt.title('Most Popular categories',fontsize=20, fontweight='bold')

plt.show()
```

Most popular categories of Netflix content are shown on this chart.

```python
fig = px.bar(x = netflix_data[netflix_data['type']=='TV Show']['duration'].value_counts().index,
            y = netflix_data[netflix_data['type']=='TV Show']['duration'].value_counts())

fig.update_layout(
    title='Duration of TV Shows',
    xaxis_title="Duration (seasons)",
    yaxis_title="Total number",
    xaxis = dict(
        tickmode = 'linear',
        tick0 = 1,
        dtick = 1)
)

fig.show()
```

Most TV shows on Netflix have no more than one season.

```python
plt.figure(figsize=(12,10))
ax = sns.countplot(x="rating", data=netflix_data, order=netflix_data['rating'].value_counts().index[0:14])
plt.title('Most popular rankings of Netflix content', pad = 15, fontsize = 18)
```

Most of the content of Netflix is either rated TV-MA or TV-14

In conclusion, analysis of the data set gave information about TV shows and movies on Netflix.
AS we can see, there are more movies on Netflix than TV Shows. The most popular rating that Netflix content has is TV-MA. Also, 2018 was the year with the most releases, and the USA is the main country that does content for Netflix.
